```python
from datetime import datetime
import time

def read_record(record):
    record_date = record[1:record.find(']')]
    dt = datetime.strptime(record_date, "%Y-%m-%d %H:%M")
    record_text = record[record.find(']') + 2:]
    guard_id = None
    if '#' in record_text:
        guard_id = record_text[record_text.find('#') + 1:record_text.find(' ', record_text.find('#'))]
    state = 'start'
    if 'falls asleep' in record_text:
        state = 'sleep'
    if 'wakes up' in record_text:
        state = 'awake'
    return {'time':dt, 'data':record_text, 'guard': guard_id, 'state':state}
    
an_example = '[1518-11-01 23:58] Guard #99 begins shift'
read_record(an_example)
```

```python
import pandas as pd

def get_ordered_list(input):
    data = []
    for row in input.split('\n'):
        if row == '':
            continue
        else:
            data.append(read_record(row))
            
    data.sort(key=lambda x: x['time'])
    return data

print(get_ordered_list(example_input))
```
