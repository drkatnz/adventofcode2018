```python
def read_claim(claim_text):
    blocks = claim_text.split()
    claim_id = blocks[0][1:]
    margin_left = int(blocks[2][:blocks[2].find(',')])
    margin_top = int(blocks[2][blocks[2].find(',') + 1:-1])
    width = int(blocks[3][:blocks[3].find('x')])
    height = int(blocks[3][blocks[3].find('x') + 1:])
    return claim_id, margin_left, margin_top, width, height

def get_max_width(claim):
    return claim[1] + claim[3]

def get_max_height(claim):
    return claim[2] + claim[4]
    
example_claim = '#123 @ 3,2: 5x4'
claim = read_claim(example_claim)
print(claim)
print(get_max_width(claim))
print(get_max_height(claim))
```

```python
def get_dimensions(claim_set):
    max_width = -1
    max_height = -1
    for row in claim_set.split('\n'):
        if row.strip() == '':
            continue
        claim = read_claim(row)
        if get_max_width(claim) > max_width:
            max_width = get_max_width(claim)
        if get_max_height(claim) > max_height:
            max_height = get_max_height(claim)

    return max_width, max_height

print(get_dimensions(example_set))
```

```python
import numpy as np

def get_overlap(claim_set):
    dimensions = get_dimensions(claim_set)
    fabric = np.zeros(dimensions)
    for row in claim_set.split('\n'):
        if row.strip() == '':
            continue
        claim = read_claim(row)
        for x in range(claim[3]):
            for y in range(claim[4]):
                fabric[claim[1] + x][claim[2] + y] = fabric[claim[1] + x][claim[2] + y] + 1
                
    total_overlap = 0
    for x in range(dimensions[0]):
        for y in range(dimensions[1]):
            if fabric[x][y] > 1:
                total_overlap = total_overlap + 1
    return total_overlap
print(get_overlap(example_set))
```

```python
def get_non_overlap_claim(claim_set):
    dimensions = get_dimensions(claim_set)
    fabric = np.zeros(dimensions)
    for row in claim_set.split('\n'):
        if row.strip() == '':
            continue
        claim = read_claim(row)
        for x in range(claim[3]):
            for y in range(claim[4]):
                fabric[claim[1] + x][claim[2] + y] = fabric[claim[1] + x][claim[2] + y] + 1
                
    for row in claim_set.split('\n'):
        if row.strip() == '':
            continue
        claim = read_claim(row)
        no_overlaps = True
        for x in range(claim[3]):
            for y in range(claim[4]):
                if(fabric[claim[1] + x][claim[2] + y] > 1):
                    no_overlaps = False
                    break
            if no_overlaps == False:
                break
        if no_overlaps:
            return claim[0]
    
print(get_non_overlap_claim(example_set))
print(get_non_overlap_claim(actual_input))
```
